```python
# Load datasets (Iris & California Housing)
from sklearn.datasets import load_iris, fetch_california_housing
import pandas as pd

# Iris dataset (Classification)
iris = load_iris(as_frame=True)
X_iris, y_iris = iris.data, iris.target
iris_df = iris.frame

print("Iris Dataset:")
display(iris_df.head())

# California Housing dataset (Regression)
housing = fetch_california_housing(as_frame=True)
X_housing, y_housing = housing.data, housing.target
housing_df = housing.frame
housing_df['MedHouseVal'] = y_housing

print("California Housing Dataset:")
display(housing_df.head())
```

```python
# Train-test split
from sklearn.model_selection import train_test_split

# Iris split
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42)

# Housing split
X_train_housing, X_test_housing, y_train_housing, y_test_housing = train_test_split(
    X_housing, y_housing, test_size=0.2, random_state=42)
```

```python
# Train models
from sklearn.linear_model import LogisticRegression, LinearRegression

# Logistic Regression (Iris)
iris_model = LogisticRegression(max_iter=200)
iris_model.fit(X_train_iris, y_train_iris)
iris_preds = iris_model.predict(X_test_iris)

# Linear Regression (Housing)
housing_model = LinearRegression()
housing_model.fit(X_train_housing, y_train_housing)
housing_preds = housing_model.predict(X_test_housing)
```

```python
# Evaluation for Iris (Classification)
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score

iris_acc = accuracy_score(y_test_iris, iris_preds)
iris_cm = confusion_matrix(y_test_iris, iris_preds)
iris_precision = precision_score(y_test_iris, iris_preds, average="macro")
iris_recall = recall_score(y_test_iris, iris_preds, average="macro")

print("Iris Model Evaluation")
print("Accuracy:", iris_acc)
print("Confusion Matrix:\n", iris_cm)
print("Precision:", iris_precision)
print("Recall:", iris_recall)
```

```python
# Evaluation for Housing (Regression)
from sklearn.metrics import mean_squared_error
import numpy as np

housing_rmse = np.sqrt(mean_squared_error(y_test_housing, housing_preds))

print("Housing Model Evaluation")
print("RMSE:", housing_rmse)
```
